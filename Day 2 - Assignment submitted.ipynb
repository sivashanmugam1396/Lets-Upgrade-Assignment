# ASSIGNMENT - 1

```python
eng = 75
math = 80
sci = 66
soc = 78
```

```python
tot = eng + math + sci + soc
avg = tot/4
```

```python
if avg>90:
    grade = 'A'
elif avg<=90 and avg>80:
    grade = 'B'
elif avg<=80 and avg>70:
    grade = 'C'
elif avg<=70 and avg>60:
    grade = 'D'
elif avg<=60 and avg>50:
    grade = 'E'
elif avg<50:
    grade = 'F'
```

```python
print("total : ", tot)
print("average : ", avg)
print("Grade : ", grade)
```

```python
Number = int(input(" Please Enter any Number: "))

print("Prime numbers between", 1, "and", Number, "are:")

for num in range(1, Number + 1):
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num)
```
